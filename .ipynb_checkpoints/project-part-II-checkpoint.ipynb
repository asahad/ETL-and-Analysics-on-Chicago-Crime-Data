### 3- Construct a query to retrieve the count of case incidents, count and percent (rounded to 3 decimal places) of arrests for each year;  order descending by arrested percent.
<span style="font-size:7px"><b>Hint:</b> 22 rows affected -- data[2015=26.451%] -- parse dates, uses "case when"</span>

```python
%%sql
SELECT 
EXTRACT(YEAR FROM incident_date) AS year, 
COUNT(*) AS num_incidents, 
SUM(CASE WHEN arrest=1 THEN 1 ELSE 0 END) AS num_arrests, 
ROUND(SUM(CASE WHEN arrest=1 THEN 1 ELSE 0 END) ::decimal  / COUNT(*)::decimal, 3) AS arrest_percent
FROM cc_cases
GROUP BY year
ORDER BY arrest_percent DESC;
```

### 5- For each year, which month is ranked #1 as having the greatest number of crime incidents with a primary crime description containing the term "NARCOTICS" or secondary crime desciption containing the phrase "GUN"? Display the year, month of the cases incident date, the primary and secondary crime description, the count of incidents and the ranking number. Order by year descending.

<span style="font-size:7px"><b>Hint:</b>22 rows affected -- data[2012-2, incidents=1942] ranking# should all be 1 -- involves parsing dates, a nested query with a window function and groups</span>

```python
%%sql
SELECT 
EXTRACT(YEAR FROM incident_date) AS year, 
EXTRACT(MONTH FROM incident_date) AS month,
iucr_primary_desc,
iucr_secondary_desc,
COUNT(*) AS num_incidents,
RANK() OVER (PARTITION BY EXTRACT(YEAR FROM incident_date) ORDER BY COUNT(*) DESC) AS ranking
FROM cc_cases
JOIN cc_iucr_codes_primary_descriptions ON cc_cases.iucr_code=cc_iucr_codes_primary_descriptions.iucr_code
JOIN cc_iucr_codes_secondary_descriptions ON cc_cases.iucr_code=cc_iucr_codes_secondary_descriptions.iucr_code
WHERE iucr_primary_desc LIKE '%NARCOTICS%' OR iucr_secondary_desc LIKE '%GUN%' AND ranking=
GROUP BY year, month, iucr_primary_desc, iucr_secondary_desc
ORDER BY year DESC;
```

### 6 What is the average difference (in days, expressed as an integer) between the updated and incident dates for case arrests and primary and secondary crime descriptions ordered by the average difference in days.

<span style="font-size:7px"><b>Hint:</b>HINT: 538 rows affected -- output first row data[NARCOTICS,SALE/DEL DRUG PARAPHERNALIA,False,6] -- Involves Aggregate groups, parse date, calculate date difference</span>

```python
%%sql
SELECT 
iucr_primary_desc,
iucr_secondary_desc,
ROUND(AVG(updated_on::date-incident_date::date)) AS avg_diff_days
FROM cc_cases
JOIN cc_iucr_codes_primary_descriptions ON cc_cases.iucr_code=cc_iucr_codes_primary_descriptions.iucr_code
JOIN cc_iucr_codes_secondary_descriptions ON cc_cases.iucr_code=cc_iucr_codes_secondary_descriptions.iucr_code
WHERE arrest = 1
GROUP BY iucr_primary_desc, iucr_secondary_desc
ORDER BY avg_diff_days;
```

**One potential query that could be useful for policing planning, policy making, and citizen awareness in PostgreSQL is: "What are the most common crime types in the city of Chicacgo, and how have their rates changed over the last year?" This query is be written as follows:**

```python
%%sql
SELECT distinct iucr_secondary_desc,
       ROUND(AVG(crime_count) OVER (PARTITION BY iucr_secondary_desc)) AS avg_crime_count
FROM (SELECT to_char(incident_date, 'Mon') AS month,
             iucr_secondary_desc,
             COUNT(*) AS crime_count
      FROM cc_cases 
      JOIN cc_iucr_codes_secondary_descriptions ON cc_cases.iucr_code=cc_iucr_codes_secondary_descriptions.iucr_code
      WHERE incident_date >= current_date - interval '1 year'
      GROUP BY month, iucr_secondary_desc) AS monthly_crimes
GROUP BY iucr_secondary_desc,monthly_crimes.crime_count
ORDER BY avg_crime_count DESC
```

**Query Explanation**

This query would first group the crime reports using their secondary description  by month and crime type, then calculate the average number of crimes for each crime type over the last year. It would then order the crime types by their average crime count, so that the most common crime types are displayed first. By using a window function (AVG() OVER (PARTITION BY iucr_secondary_desc)), we can also easily compare the average crime counts for each crime type. This information could be useful for law enforcement agencies to identify trends in criminal activity and allocate resources accordingly. It could also help policy makers develop strategies to address the most common types of crimes in the city. Additionally, providing this information to the public could help raise awareness about crime in the area and encourage citizens to take steps to protect themselves and their communities. For example over the past one year there is an average of 1101 reported crimes to vehicles per month followed by 943 reported crime to property.This will ensure city authorities as well as residents know which crimes are more prevalent and take steps to reduce them.
